```python
#generate the descriptions for every transcation in dataset manually
import pandas as pd
df = pd.read_csv('test_data(copy).csv') 
df.head()
```

```python
for index, row in df.iterrows():
    sentence = 'Transaction no ' + str(row['Transaction_Number']) + ' is a '
    if(row['Transaction_type']=='Credit'):
        sentence = sentence + 'credit transaction,'
    else:
        sentence = sentence + 'debit transaction,'
    if row['Source_Country'] != row['Dest_Country']:
        sentence = sentence + ' from ' +row['Source_Country'] + ' to ' + row['Dest_Country'] + ' worth ' + str(row['Amount'])
    else:
        sentence = sentence + ' within ' + row['Source_Country'] + ' worth ' + str(row['Amount'])
    currency = row['Currency']
    if currency == 'USD':
        sentence = sentence + ' dollars, which is '
    elif currency == 'INR':
        sentence = sentence + ' ruppees, which is '
    elif currency == 'Euro':
        sentence = sentence + ' euros, which is '
    else:
        sentence = sentence + ' yen, which is '
    if row['Is_suspicious'] == 0:
        sentence = sentence + 'suspicious with ' + str(row['Occurance_probability']) + ' probability.'
    else:
        sentence = sentence + 'not suspicious with ' + str(row['Occurance_probability']) + ' probability.'
    print(sentence)
    with open('description.txt', 'a') as the_file:
        the_file.write(sentence + '\n')
```
